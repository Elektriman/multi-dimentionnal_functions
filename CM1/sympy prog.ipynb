# Tests d'utilisation de sympy pour le calcul de dérivées

```python
from sympy import *
```

```python
from sympy.plotting.plot import plot3d

x, y, z = symbols('x y z')
f = exp(x) + exp(2*y) + exp(3*z)
gradF = diff(f, x) + diff(f, y) + diff(f, z)
plot3d(f.replace(z,4), (x,-1,1), (y,-1,1))
```

```python
plot3d(gradF.replace(z,4), (x,-1,1), (y,-1,1))
```
