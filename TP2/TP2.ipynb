### 1.1. Lignes brisées

Une ligne brisée relie des points donnés. On donne donc la liste des abscisses, la liste
des ordonnées et on effectue le tracé :

```python
x = [1., 2.5, 4.]
y = [3., 1., 5.]
plt.axis('equal')
plt.plot(x,y)
plt.axis([-1., 5., -1., 6.])
plt.grid()
plt.show()
```

L'option equal permet d'avoir les mêmes échelles sur les deux axes, `pyplot.axis` (ou `plt.axis` puisqu'on a défini le nom plt pour taper plus vite) permet de tracer les axes en choisissant la fenêtre d'affichage, `pyplot.grid` donne la grille. Et il faut toujours terminer par `plot.show` pour que Python nous montre l'oeuvre finale.

La fonction plot a de nombreuses options de présentation. Le paramètre color permet de choisir la couleur (`g` : vert, `r` : rouge, `b` : bleu). Pour définir le style de la ligne, on utilise linestyle (`-` : ligne continue, `--` : ligne discontinue, `:` : ligne pointillée...). Si on veut marquer les points des listes, on utilise le paramètre marker (`+`, `.`, `o`, `v` donnent différents symboles).

Essayez par exemple :

```python
x = [1., 2.5, 4.]
y = [3., 1., 5.]
plt.axis([-1., 5., -1., 6.])
plt.plot(x, y, color='r', linestyle=':', marker='o')
plt.show()
```

Il y a aussi l'option linewidth pour choisir l'épaisseur des traits (essayez par exemple `linewidth=3.5, linestyle="-"` dans un tracé), l'instruction `pyplot.figure` pour déformer la taille de la fenêtre d'affichage (essayez par exemple `plt.figure(figsize=(10,6), dpi=80)` et `plt.figure(figsize=(8,6), dpi=80)` dans un tracé).

```python
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x), color="red", linewidth=2.5, linestyle="-", label="sinus")
plt.plot(x, np.cos(x), color="blue", linewidth=2.5, linestyle="-", label="cosinus")
plt.legend(loc='upper left')
plt.axis([-np.pi, np.pi,-1,1])
plt.yticks([-1, 0, 1],[r'$-1$',r'$0$',r'$1$'])
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-{\pi / 2}$', r'$0$', r'$\pi / 2$', r'$\pi$'])
```

• `$\pi$ ` : on met des symboles `$` pour pouvoir utiliser des caractères mathématiques spéciaux tel que `\pi` qui donne $\pi$ en Latex.

C'est typiquement le genre d'exercice torture quand on vous oblige à faire à la main (double tableau de variation etc). Avec Python, on calcule d'abord simplement la liste des valeurs prises par le paramètre puis on construit la liste des abscisses et des ordonnées correspondantes.

### 1.4. Tracé d'une surface

Pour tracer une surface en 3D, disons d'équation $z = f(x, y)$, on définit d'abord une grille du plan, en $(x, y)$ donc, puis on calcule toutes les valeurs de $z = f(x, y)$ correspondantes. Puis on affiche. <br>

```python
ax = Axes3D(plt.figure())
def f(x,y):
    return x**2 - y**2
X = np.arange(-1, 1, 0.02)
Y = np.arange(-1, 1, 0.02)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
ax.plot_surface(X, Y, Z)
plt.show()
```

Reproduisez et amliorez (pour y voir plus net ! il serait bon en particulier de pouvoir clairement les valeurs prises par la fonction le long des lignes de niveau et/ou le long des vecteurs...) et expliquez ce qui est représenté. Quel rapport avec l'exercice sur le capitaine Kirk ?

$f(x,y) = e^{-x} + e^{-2y}$
\begin{align*}
    f(x,y) = k &\Longleftrightarrow e^{-x} + e^{-2y} = k \\
    &\Longleftrightarrow e^{-2y} = k - e^{-x} \\
    &\Longleftrightarrow -2y = ln (k - e^{-x}) \\
    &\Longleftrightarrow y = -{1 \over 2}ln (k - e^{-x})
\end{align*}
$g(x) = -{1 \over 2}ln (k - e^{-x})$ est définie sur l'intervalle où $k - e^{-x} > 0$
\begin{align*}
    k - e^{-x} > 0 &\Longleftrightarrow e^{-x} > k \\
    &\Longleftrightarrow -x > ln(k) \\
    &\Longleftrightarrow x > -ln(k) \\
\end{align*}

```python
f = lambda x,y : np.exp(-x) + np.exp(-2*y)
x2 = np.linspace(-3, 3, 31)
y2 = np.linspace(-2, 2, 31)
X2, Y2 = np.meshgrid(x2, y2)
Z2 = f(X2, Y2)
#d_xf, d_yf = np.gradient(Z2)

#fonction qui renvoie un gradient local de la fonction en P avec un intervalle de h
#alors df/dx ~= (f(x+h)-f(x-h))/2h
def Grad(f, P:tuple, h:float)->tuple:
    x,y = P
    dx = (f(x+h,y)-f(x-h,y))/(2*h)
    dy = (f(x,y+h)-f(x,y-h))/(2*h)
    return dx,dy

fig, ax = plt.subplots(1,1, figsize=(15,10))

C = ax.contour(X2, Y2, Z2, levels=[3,5,10,20,30,50]) #tracé du contour
for l in C.collections : #on va chercher les éléments tracés par la fonction
    for t in l.get_segments(): #on va chercher toutes les lignes tracées point par point
        dx, dy = Grad(f, (t[:,0], t[:,1]), 0.001) #on calcule le gradient autour d'un point
        
        #on affiche la flèche de gradient sur chaque point tracé par la fonction contour
        ax.quiver(t[:,0], t[:,1], dx, dy, color='r', angles='xy', scale=1000, alpha = 0.5, width=0.003)
ax.grid()
plt.show()
```

## 3. Exercice 2 : un autre exercice pour se souvenir de ce que donne le gradient géométriquement, création d'un filtre de Sobel pour le traitement d'image

On vient de voir que le gradient donne la direction de plus grande croissance d'une fonction. Une autre façon
de le dire est que le gradient est un vecteur perpendiculaire au plan tangent à la surface :

<img src="img/gradient.png" alt="alt_text" align="left"/>

Dans cet exercice, on va utiliser un autre point de vue sur le vecteur gradient. On ne va plus regarder sa
direction mais sa norme. Si cette norme est grande, cela signifie que la valeur absolue de l'une au moins des
dérivées partielles est grande et que donc, chaque dérivée étant la limite d'un taux de variation, la surface
représentative de la fonction présente, dans une direction au moins, de grandes variations.<br>
Nous allons exploiter cette propriété pour créer un programme de détection de contours dans une image.
Certes il est rustique mais en une dizaine de lignes de programme, il va nous permettre d'extraire les
"contours" dans une image et d'obtenir ce genre de choses :

correspondent aux brusques variations de la luninance. Si on fait une cartographie 3D de l'image, i.e. si on

on devine que les contours correspondent aux zones où la norme du gradient de $l$ est très, très grande (les
zones de décroissance ou croissance très, très abruptes de $l$). La détection de contour va dès lors consister
à créer une image en noir et blanc où ne seront noirs que les pixels correspondant à des points où la norme
du gradient de $l$ est grande.

Des indications :
1. Un module spécifique au traitement d'image est PIL. Ici, vous n'aurez besoin a priori que de charger le sous-module Image : `from PIL import Image`.

2. Pour ouvrir le fichier avec l'image, disons `machin.truc` avec `truc = jpg` ou `truc = png` etc, utilisez l'instruction `Image.open`. SI vous appelez cet objet `imageDepart`, cela donne : <br><center>`imageDepart = Image.open (machin.truc)`</center> <br>
Vous pouvez afficher une image, par exemple celle que vous avez appelée ___imageTruc___, avec : <br><center>`imageTruc.show()`</center><br>
Pour connaître la taille de l'image que vous venez de charger, nombre de colonnes et de nombre de lignes (une image n'est qu'un tableau de chiffres...), faites<br><center> `NbColonnes,NbLlignes = imageDepart.size`</center>

3. Pour récupérer dans `imageDepart` l'information portée par un pixel de coordonnées $(i, j)$, avec $(i, j) \in \mathbb{N}^2, 0 \le i \le NbLignes$ et $0 \le j \le NbColonnes$, faites :<br><center>`v = imageDepart.getpixel((j,i))`</center>
(et oui, le numéro de colonne en premier). Pour une image en couleur, vous récupérez ainsi un vecteur v de R3, avec : <br>
$v[0]$=quantité de rouge portée par le pixel,<br>
$v[1]$=quantité de vert portée par le pixel et<br>
$v[2]$=quantité de bleu portée par le pixel <br>
(c'est du RGB, Red Green, Blue ; une image en niveaux de gris ne donnerait qu'une valeur, la luminance). Ces quantités varient toutes de 0 à 255.<br>
Un pixel est noir si le vecteur v correspondant est :
<br><center>v =( 255, 255, 255 )</center>
Pour au contraire attibuer une valeur v à un pixel de l'image ___imageTruc___:
<br><center>`imageTruc.putpixel((j,i),v)`</center>
4. Pour créer une autre image, par exemple de mêmes type et taille que l'image de départ :
<br><center>`image2 = Image.new( imageDepart.mode, imageDepart.size)`</center>
5. Calculer un gradient sur des données discrètes est finalement facile puisque chaque dérivée partielle est alors un taux de variation avec un pas $h = 1$.
6. Pour créer une fonction qui calcule la norme d'un vecteur de R2 vous n'avez pas besoin de moi, ni pour l'utiliser pour calculer la norme du gradient.
7. Vous pouvez aussi chercher les contours "essentiellement verticaux" ou "essentiellement horizontaux" très facilement non ?

```python
from PIL import Image

def filtre(img, w=1):
    res_img = Image.new(img.mode, img.size) #création de l'image à rendre
    C, L = img.size
    I = np.array(img.getdata()).reshape(L, C, 3) #conversion en tableau numpy
    m = np.array((min(I.ravel()),)*3) #récupération du min
    M = np.array((max(I.ravel()),)*3) #récupération du max
    I = (I-m)/M #normalisation des valeurs de l'image
    Grad = np.zeros((L,C,2,3)) #création du tableau de gradient 

    for i in range(L):
        for j in range(C):

            #pour chaque pixel i,j on cherche quels sont les deux pixels qui vont être utilisés pour le calcul de gradient
            #soit celui qui est avant et celui qui est après, verticalement ou horizontalement
            x1,x2 = None,None
            y1,y2 = None,None
            if i==0 :
                x1 = I[i,j]
            else :
                x1 = I[i-1,j]

            if j==0 :
                y1 = I[i,j]
            else :
                y1 = I[i,j-1]

            if i==L-1 :
                x2 = I[i,j]
            else :
                x2 = I[i+1,j]

            if j==C-1 :
                y2 = I[i,j]
            else :
                y2 = I[i,j+1]

            Grad[i,j][0],Grad[i,j][1] = (x2-x1)/2 , (y2-y1)/2 #calcul du gradient
            n = (Grad[i,j,0]**2 + Grad[i,j,1]**2)**0.5 #calcul de la norme du gradient
            n = n**w #ajustement du filtre
            res_img.putpixel((j,i), tuple(np.array(n*256, dtype=int))) #ajout du pixel correspondant sur l'image résultat
    
    return res_img
    
if __name__ == "__main__" :
    img = Image.open("img/carrés.png") #ouverture de l'image
    w = 0.8 #valeur à modifier pour accentuer l'effet du filtre. plus la valeur est proche de 0, 
    #plus il nous sera possible de voir la différence entre deux couleurs proches.
    L = [] #liste pour faire plusieurs exemples de l'accentuation du filtre
    for w in np.linspace(1,0.5,3):
        im = filtre(img, w)
        L.append(np.array(im).reshape(im.size[1],im.size[0],3))
    
    #affichage de tous les résultats sur une figure avec 3 lignes, 1 colonne
    fig,ax = plt.subplots(3,1, figsize=(15,30))
    for i in range(3):
        ax[i].imshow(L[i])
        ax[i].set_title(f"w = {np.linspace(1,0.5,3)[i]}")
```
