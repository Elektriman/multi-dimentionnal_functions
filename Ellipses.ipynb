# Ellipses et cercles

```python
import numpy as np
from matplotlib import patches
import matplotlib.pyplot as plt
```

```python
a = 3
b = 2

fig = plt.figure()
ax = fig.add_subplot()

e2 = patches.Ellipse((0, 0), 2*a, 2*b, linewidth=2, fill=False, zorder=2)

ax.add_patch(e2)
ax.grid()
ax.axis('equal')
```

```python
def give_function(a,b):
    return lambda x : np.sqrt(b**2 - (x*b/a)**2)

f = give_function(a,b)
X = np.linspace(0,a,1000)
plt.plot(X,f(X))
plt.axis('equal')
```

```python
T = np.array([X, f(X)]).transpose()
theta = np.array([np.arccos(T[i,0]/a) for i in range(len(X))])

plt.plot(X, theta)
```

```python
a = 3
b = 2

fig = plt.figure()
ax = fig.add_subplot()

e2 = patches.Ellipse((0, 0), 2*a, 2*b, linewidth=2, fill=False, zorder=2)

ax.add_patch(e2)
ax.grid()
ax.axis('equal')


x1 = 2.7
y1 = f(x1)
span = np.linspace(-0.3,0.3,30)
ax.plot(x1+span*(-y1*a**2), y1 + span*(x1*b**2), ls='--')
ax.plot(x1+span*(x1*b**2), y1 + span*(y1*a**2), ls='--')
ax.scatter(x1 * (1 - b**2/a**2) ,0)
c1 = patches.Circle((x1 * (1 - b**2/a**2) ,0), np.sqrt((x1 * (b**2/a**2))**2 + y1**2), fill=False)
ax.add_patch(c1)
```
