(a) $f(x,y)=\frac{x^2+y^2+2xy}{y^4}$

Soit $k \in \mathbb{R}$
\begin{align*}
    f(x,y) = k &\Longleftrightarrow \frac{x^2+y^2+2xy}{y^4} = k\\
    &\Longleftrightarrow \frac{(x+y)^2}{y^4} = k \\
    &\Longleftrightarrow \frac{x+y}{y^2} = \pm \sqrt{k} \\
    &\Longleftrightarrow x+y = \pm \sqrt{k}y^2 \\
    &\Longleftrightarrow \pm \sqrt{k}y^2+y+x = 0 \\
\end{align*}

Soit $P$ le polynôme $-\sqrt{k}X^2 + X + x$<br>
Soit $Q$ le polynôme $\sqrt{k}X^2 + X + x$<br>
$\Delta_P = 1-4(-\sqrt{k})x = 1+4\sqrt{k}x$<br>
$\Delta_Q = 1-4\sqrt{k}x$<br>
Pour $x < -{1 \over 4\sqrt{k}}$, P ne s'annule jamais, mais Q s'annule.<br>
Pour $x > \phantom{-}{1 \over 4\sqrt{k}}$, Q ne s'annule jamais, mais P s'annule.<br>
Donc il existe toujours une solution pour y<br>

$r^+_P = \frac{1+\sqrt{1+4\sqrt{k}x}}{2\sqrt{k}}$ et $r^-_P = \frac{1-\sqrt{1+4\sqrt{k}x}}{2\sqrt{k}}$<br>
$r^+_Q = \frac{-1-\sqrt{1-4\sqrt{k}x}}{2\sqrt{k}} = -r^+_P$ et $r^-_Q = \frac{-1+\sqrt{1-4\sqrt{k}x}}{2\sqrt{k}} = -r^-_P$<br>
or $r_p^-(x) = r_P^+(-x)$<br>
donc $\forall x \in \mathbb{R}^+$, $y = \pm r_P^+(\pm x) = \pm \frac{1 + \sqrt{1 \pm 4\sqrt{k}x}}{2\sqrt{k}}$

```python
x = np.linspace(-3,3,100)
f = lambda k : (lambda x: (1+np.sqrt(1+4*np.sqrt(k)*x)) / 2*np.sqrt(k) )
colors = plt.cm.viridis(np.linspace(0,1,5))
for k in range(1,5):
    plt.plot(x, f(k)(x), c = colors[k])
    plt.plot(x, f(k)(-x), c = colors[k])
    plt.plot(x, -f(k)(x), c = colors[k])
    plt.plot(x, -f(k)(-x), c = colors[k])
```

(b) $f(x,y) = ln(x^2 + y^2)$

Soit $k \in \mathbb{R}$
\begin{align*}
    f(x,y) = k &\Longleftrightarrow ln(x^2 + y^2) = k\\
    &\Longleftrightarrow x^2 + y^2 = e^k\\
    &\Longleftrightarrow y^2 = e^k - x^2 \\
    &\Longleftrightarrow y = \pm \sqrt{e^k - x^2} \\
\end{align*}

```python
x = np.linspace(-3,3,1000)
f = lambda k : (lambda x: np.sqrt(np.exp(k)-x**2))
colors = plt.cm.viridis(np.linspace(0,1,5))
values = np.linspace(-2,2,5)
for i in range(5):
    plt.plot(x, f(values[i])(x), c=colors[i])
```
